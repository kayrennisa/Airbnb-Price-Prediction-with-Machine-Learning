
# Multiple Linear Regression


```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model

from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
```

## E.1(a) - Listing data without Feature Selection

```python
# IMPORT LISTING DATA
listings1 = pd.read_csv('a.csv')
listings1.head()
```

```python
# DROP UNNECCESSARY COLUMNS
listings1 = listings1.drop(columns='Unnamed: 0')
listings1 = listings1.drop(columns='id')
listings1.head()
```

```python
X = listings1.drop('price',axis=1)
y = listings1['price']
```

```python
# Prepare the train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# construct 
mlr = linear_model.LinearRegression()
```

```python
# Fit the classifier
mlr = mlr.fit(X_train, y_train)
```

```python
y_pred = mlr.predict(X_test)
```

```python
scores = cross_val_score(mlr, X_train, y_train, cv=5)
```

```python
# Calculate performance metrics
mlr1_dict = {'Model':'MLR (a)',
          'R^2':metrics.r2_score(y_test, y_pred),
          'MAE':metrics.mean_absolute_error(y_test, y_pred),
          'MSE':metrics.mean_squared_error(y_test, y_pred),
          'RMSE':np.sqrt(metrics.mean_squared_error(y_test, y_pred)),
          'CVS':scores.mean()
          }
```

```python
# Display model performance metrics
mlr1_metrics = pd.DataFrame.from_dict(mlr1_dict, orient = 'index').T
mlr1_metrics
```
